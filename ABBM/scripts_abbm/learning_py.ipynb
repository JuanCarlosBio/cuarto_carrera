```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
practica_2 = pd.read_csv("https://raw.githubusercontent.com/Juankkar/cuarto_carrera/main/ABBM/data/practica_6.csv")
columnas = ['alumno','grupo','blanco','tubo1','tubo2','tubo3','tubo4','tubo5','tubo6','tubo7',
            'albu_blanco','albu1','albu2','albu3','albu4','albu5']
practica_2.columns=columnas
practica_2.head(n=6)
```

```python
# Mi primer :) plot de python, dios que Noob xd

plt.scatter(practica_2["tubo1"], practica_2["tubo2"])
plt.title("Absorbacias para los tubos 1 y 2 de todos los grupos")
plt.xlabel("Absorbancias del Tubo 1")
plt.ylabel("Absorbancias del tubo 2")
```

```python
# Seleccinar columnas, en este caso no me interesan las muestras de la 
# albúmina.

tubos = practica_2[['alumno', 'grupo', 'blanco', 'tubo1',
                   'tubo2','tubo3','tubo4', 'tubo5','tubo6','tubo7']]                 
tubos.head(n=5)
```

```python
# Filtrar datos, en este caso los alumnos del grupo 1 y que los valores
# del tubo1 sean mayores a 0.15
tubos[(tubos['grupo'] == 'Grupo 1') & (tubos['tubo1'] > .15)] 
```
