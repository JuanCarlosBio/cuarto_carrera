### **Variables (grupos de los ácidos grasos) que son normales y pero no homocedásticos** 

Todas las variables fuereno significativas en el test de Welch

````python
# Importamos las librarías necesarias
import pandas as pd
import numpy as np
import scikit_posthocs as sp
````

````python
# Importamos los datos de las artemias
matrix_artemias = pd.read_csv("https://raw.githubusercontent.com/Juankkar/cuarto_carrera/main/FAA/artemias_sripts_datos/matrix_artemias.csv")
rca2=pd.read_csv("https://raw.githubusercontent.com/Juankkar/cuarto_carrera/main/FAA/artemias_sripts_datos/rca.csv")
matrix_artemias.head(n=6)
````

````python
# Hacemos la condición para cambiar los grupos de Tratamiento
condicion = [
    (matrix_artemias["Tratamiento"] == 1),
    (matrix_artemias["Tratamiento"] == 2),
    (matrix_artemias["Tratamiento"] == 3),
    (matrix_artemias["Tratamiento"] == 4),
]
# Nomres de los grupos
nombres = ["Levadura", "Lectina marian", "Echium/bacalao", "Enriquecedor comercial"]
# Cambio mediante la función select de numpy
matrix_artemias["Tratamiento"] = np.select(condicion, nombres)
matrix_artemias.head(n=6)
# OJO!!! Si le das a correr una segunda vez se vuelve loco y llama a todos los grupos 0, tendriás que correr de nuevo el bloque de arriba y este
# para solucionarlo.
````

````python
# Hacemos Tamhane T2.
# AGs, normales pero no homocedásticos homocedásticos:
# C 16:1n-7; C 18:0; C 18:1n-9; C 18:3n-6; C 20:5n-3; C 22:6n-3
def resumen(acido_graso):
    cuerpo = sp.posthoc_tamhane(matrix_artemias, val_col=acido_graso, group_col='Tratamiento', welch=True)
    return(cuerpo)


C_16_1n7= resumen('C 16:1n-7')
C_18_0 = resumen('C 18:0')
C_18_1n9= resumen('C 18:1n-9')
C_18_3n6= resumen('C 18:3n-6')
C_20_5n3= resumen('C 20:5n-3')
C_22_6n3= resumen('C 22:6n-3')
````

### ***Resultados de la RC2 de las prácticas de informática***

````python
# Inferencia estadística de la componente 2 del pca rotado:
RC2_artemias= sp.posthoc_tamhane(rca2, val_col='valores', group_col='tratamiento', welch=True)
RC2_artemias
````
